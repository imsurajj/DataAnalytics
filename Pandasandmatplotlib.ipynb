```python
import numpy as np
```

```python
a = np.array([1, 2, 3])
b = np.array([[1, 2], [3, 4]])
print(a)
print(b)
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

print(arr.shape)     # (2, 3)
print(arr.size)      # 6
print(arr.ndim)      # 2
print(arr.dtype)     # int64
```

```python
a = np.array([1, 2, 3, 4, 5, 6])
print(a.reshape(2, 3))
```

```python
arr = np.array([10, 20, 30, 40, 50])
print(arr[1])       # 20
print(arr[1:4])     # [20 30 40]
```

```python
print(np.zeros((2, 3)))         # 2x3 array of 0
print(np.ones((3, 3)))          # 3x3 array of 1
print(np.eye(3,3))                # Identity matrix
print(np.full((2, 2), 7))       # Filled with 7
print(np.arange(1, 10, 2))      # [1 3 5 7 9]
print(np.linspace(1, 5, 4))     # [1. 2.33 3.66 5.]
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)      # [5 7 9]
print(a * b)      # [4 10 18]
print(a ** 2)     # [1 4 9]
print(np.sqrt(a)) # [1. 1.41 1.73]
print(np.mean(a)) # 2.0
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

print(np.sum(arr))          # 21
print(np.min(arr))          # 1
print(np.max(arr))          # 6
print(np.sum(arr, axis=0))  # [5 7 9] (column-wise)
print(np.sum(arr, axis=1))  # [6 15]  (row-wise)
```

```python
arr = np.array([4, 1, 5, 2])
print(np.sort(arr))  # [1 2 4 5]
```

```python
a = np.array([1, 2, 3])
b = a.copy()
c = a.view()

a[0] = 99

print(b)  # [1 2 3] (copy remains same)
print(c)  # [99 2 3] (view reflects change)
```

```python
arr = np.array([[1, 2], [3, 4]])
print(arr.flatten())  # [1 2 3 4]
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr.T)
```

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[2, 0], [1, 2]])

print(np.dot(a, b))  # Matrix multiplication
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
'total_bill':[16.99,10.34,23.68,23.68,24.59],
'tip':[1.01,1.66,3.31,3.61,4.71],
'sex':['Female','Male','Male','Male','Female'],
'smoker':['No','No','No','No','Yes'],
'day':['Sun','Sun','Sun','Sat','Fri'],
'time':['Dinner','Dinner','Dinner','Dinner','Dinner'],
'size':[2,3,3,2,4]
}

df= pd.DataFrame(data)
sns.set_style("whitegrid")

#bar plot
sns.barplot(x='day',y='total_bill',data=df)
plt.title("Barplot ")
plt.show()

#count plot
sns.countplot(x='day',data=df)
plt.title("Countplot of day")
plt.show()

#scatter plot
sns.scatterplot(x='total_bill',y='tip',data=df)
plt.title("Scatterplot of total bill and tip")
plt.show()

#Boxplot
sns.boxplot(x='day',y='total_bill',data=df)
plt.title("Boxplot ")   
plt.show()

#Violinplot
sns.violinplot(x='day',y='total_bill',data=df)
plt.title("Violinplot ")
plt.show()

#Stripplot
sns.stripplot(x='day',y='total_bill',data=df)
plt.title("Stripplot ")
plt.show()

#Swarmplot
sns.swarmplot(x='day',y='total_bill',data=df)
plt.title("Swarmplot ")
plt.show()

#pointplot
sns.pointplot(x='day',y='total_bill',data=df)
plt.title("Pointplot ")
plt.show()

#cover python 
# libraries
# data cleaning
# powerbi

```
